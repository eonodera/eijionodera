```python
import requests
import dotenv
import os

dotenv.load_dotenv()

YOUTUBE_API = os.getenv("YOUTUBE_API")
```

```python
import requests
import pprint
search_word = "Fortnite"
url = 'https://www.googleapis.com/youtube/v3/search'
response = requests.get(
    url, 
    params={
        'key': YOUTUBE_API,  # API„Ç≠„Éº„Çí„Éë„É©„É°„Éº„Çø„Å´ËøΩÂä†
        'part': 'snippet',  # ÂãïÁîªÊÉÖÂ†±„ÅÆË©≥Á¥∞„ÇíÊåáÂÆö
        'q': search_word,  # Ê§úÁ¥¢„Ç≠„Éº„ÉØ„Éº„Éâ„Çí„Éë„É©„É°„Éº„Çø„Å´ËøΩÂä†
        'type': 'video',  # ÂãïÁîª„ÅÆ„Åø„ÇíÊ§úÁ¥¢ÂØæË±°„Å´ÊåáÂÆö
        'maxResults': 10  # ÂèñÂæó„Åô„ÇãÂãïÁîª„ÅÆÊúÄÂ§ßÊï∞„ÇíÊåáÂÆö
    }
)
```

```python
# „É¨„Çπ„Éù„É≥„Çπ„ÅÆJSON„ÇíÂèñÂæó
json_data = response.json()

# JSON„ÅÆ‰∏≠Ë∫´„ÇíÁ¢∫Ë™ç
pprint.pprint(json_data)
```

```python
# „Éï„Ç£„É´„Çø„ÅÆ‰Ωø„ÅÑÊñπ
# https://developers.google.com/youtube/v3/docs/search/list?hl=ja#python%E3%80%81%E4%BE%8B-1

url = "https://www.googleapis.com/youtube/v3/search?part=snippet&maxResults=10&q=Fortnite&type=video&key=" + YOUTUBE_API

import json
res = requests.get(url)
pprint.pprint(res.json())
with open('youtube_searched.json', 'w') as f:
    json.dump(res.json(), f, indent=4)
```
